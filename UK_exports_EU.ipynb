UK exports to EU

```python
#IMPORTACION de librerias y lectura de datos
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime
filename = '/content/drive/My Drive/CODING DOJO/Códigos/Semana 3/UK_Exports_to_the_EU.xlsx'
df = pd.read_excel(filename)
df
```

```python
df_exports=df.groupby(['Month'])[[' Exports (£m)']].sum()
df_exports
```

```python
years=list(df_exports.index)
exports=list(df_exports.values)
```

```python
plt.style.use('seaborn')
plt.plot(years, exports);
plt.xlabel('Year', fontsize=16)
plt.ylabel('Exports £M', fontsize=16)
plt.ylim(bottom=0,top=22000)
plt.title('Exports from UK to EU', fontsize=18)
```

```python
df_imports=df.groupby(['Month'])[[' Imports (£m)']].sum()
df_imports
```

```python
years=list(df_imports.index)
imports=list(df_imports.values)
```

```python
plt.style.use('seaborn')
plt.plot(years, imports, c='k');
plt.xlabel('Year', fontsize=16)
plt.ylabel('Imports £M', fontsize=16)
plt.ylim(bottom=0,top=28000)
plt.title('Imports from EU to UK', fontsize=18)
```

```python
df_balance=df.groupby(['Month'])[['Balance (£m)']].sum()
df_balance
```

```python
years=list(df_balance.index)
balance=list(df_balance.values)
```

```python
plt.style.use('seaborn')
plt.plot(years, balance, c='r');
plt.xlabel('Year', fontsize=16)
plt.ylabel('Balance', fontsize=16)
plt.title('Trade Balance UK-EU', fontsize=18)
```

```python
plt.style.use('seaborn')
plt.plot(years, exports, label='Exports');
plt.plot(years, imports, c='k', label='Imports');
plt.plot(years, balance, c='r', label='Balance');
plt.legend()
```
